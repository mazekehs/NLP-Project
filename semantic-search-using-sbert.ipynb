# Semantic Search


Semantic search seeks to imporove search accuracy by understanding the content of the search query.In contrast to traditional search engines which only find documents based on lexical matches, semantic search can also find synonyms.

```python
import warnings
warnings.filterwarnings("ignore")

%pip install -U sentence-transformers
```

```python
from sentence_transformers import SentenceTransformer, util
import torch
```

```python
model=SentenceTransformer('multi-qa-MiniLM-L6-cos-v1')
```

```python
import requests
```

# Importing Data 

```python
response=requests.get('https://raw.githubusercontent.com/laxmimerit/machine-learning-dataset/5a3d98e5da8b22c9bd31430da67a59492085e887/text-dataset-for-machine-learning/sbert-corpus.txt')
corpus=response.text.split('\r\n')
response=requests.get('https://raw.githubusercontent.com/laxmimerit/machine-learning-dataset/5a3d98e5da8b22c9bd31430da67a59492085e887/text-dataset-for-machine-learning/sbert-queries.txt')
queries=response.text.split('\r\n')
```

```python
print(corpus)
```

```python
print(queries)
```

# Model Evaluation

```python
corpus_embeddings=model.encode(corpus, convert_to_tensor=True)
queries_embeddings=model.encode(queries,convert_to_tensor=True)
```

```python
#For faster calculations normalizing
corpus_embeddings=util.normalize_embeddings(corpus_embeddings)
queries_embedding=util.normalize_embeddings(queries_embeddings)
```

**Semantic Search**

```python
hits=util.semantic_search(queries_embeddings,corpus_embeddings)
```

```python
hits
```

```python
print(len(hits))
print(len(hits[0]))
```

```python
# Top 3 corpus for each query
hits=util.semantic_search(queries_embeddings, corpus_embeddings, top_k=3)
hits
```

# Scores

```python
for query, hit in zip(queries, hits):
    for q_hit in hit:
        id = q_hit["corpus_id"]
        score = q_hit["score"]
        print(query,"  <-->  ", corpus[id], "--->", score)
    print()
        
```
